```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
area = [500,600,550,700,800,850,750,500,600,650]

price = [60,65,60,68,75,80,75,58,65,67]
```

```python
area = np.array(area).reshape(-1,1)
price = np.array(price)
```

```python
LR = LinearRegression()
```

```python
LR.fit(area,price)
```

```python
price_predict  = LR.predict(area)
```

```python
price_predict
```

```python
r2 = r2_score(price,price_predict)
r2
```

```python
# to check the error
MAE = mean_absolute_error(price,price_predict)
MAE
```

```python
MSE = mean_squared_error(price,price_predict)
MSE
```

```python
RMSE = np.sqrt(mean_squared_error(price,price_predict))
RMSE
```
