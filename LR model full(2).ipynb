```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("insurance.csv")
```

```python
df.dropna(inplace = True)
```

```python
categorical_cols = ['sex', 'smoker', 'region']
encode = OneHotEncoder(drop='first',sparse_output = False)

encoded_array = encode.fit_transform(df[categorical_cols])
encoded_columns = encode.get_feature_names_out(categorical_cols)

df_encoded = pd.DataFrame(encoded_array, columns=encoded_columns)
df = df.drop(columns=categorical_cols).reset_index(drop=True)
df = pd.concat([df, df_encoded], axis=1)
```

```python
num_cols = ['age', 'bmi', 'children']
scaler = StandardScaler()
scaled_cols = pd.DataFrame(scaler.fit_transform(df[num_cols]), columns=num_cols)
```

```python
x = pd.concat([df_encoded, scaled_cols], axis=1)
y = df['charges']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("R2 Score: ", r2)
print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
```
